##### Simulate coinflips with Python!

```python
def coinflips_50():
    heads=0
    for i in range(50):
        toss = random.random()
        if toss < 0.5:
            heads+=1
    return heads

def trials(n):
    results = []
    for i in range(n):
        result = coinflips_50()
        results.append(result)
    return results
```

```python
trial_results = trials(10000)
```

```python
trial_results[:10]
```

```python
results = Counter(trial_results)
```
