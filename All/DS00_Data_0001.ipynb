#Data Transformation 
This is the technique of converting data from one format to another. 

* Data Discovery: Profiling the data to understand the structure and characteristics of data.

* Data Mapping: Defines how individual fields are mapped, modified, joined, filtered, aggregated etc. to produce the final desired output.

* Data Transformation code: Generating code to transform data based on the data mapping rules.

* Code implementation: Executing the generated code against the data to create the desired output.

* Review of data: Ensuring that the output data meets the transformation requirements. 

#Types of Data

1. Numerical: Numeric data that is quantifiable.

2. Categorical: String or non-numeric data that is qualitative in nature.

## Numerical data is further divided into the following:

* Discrete: Any numerical data that is countable. (number of people in a city). 

* Continuous: Any measurable numeric data. (height of a person or time taken)

## Categorical data is further divided into the following:

* Ordered: Any categorical data that has some order associated. (movie ratings such as  excellent, good, bad, worst) or feedback (happy, not bad, bad). 

* Nominal: Any categorical data that has no order (city names, gender, fruits).
