```python
url ="https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv"
df=pd.read_csv(url)
df
```

```python
df['decade'] = 10 * (df['year'] // 10)
df2=df.pivot_table('births', index='decade', columns='gender', aggfunc='sum')
df2
```

```python
sns.set()  # use Seaborn styles
df2.plot()
plt.ylabel('total births');
```

```python
sns.set()  # use Seaborn styles
df.pivot_table('births', index='year', columns='gender', aggfunc='sum').plot()
plt.ylabel('total births per year');
```
