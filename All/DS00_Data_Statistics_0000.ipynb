#Mean

```python
#average marks of 13 studentsc
sum=99+86+53+88+61+66+53+79+74+48+72+75+56
print("sum", sum)
print("average", sum/13)
```

```python
# Marks of Students 
#array
marks= [99,86,53,88,61,66,53,79,74,48,72,75,56]
print(marks)
```

```python
sum=0
i=0
for n in marks:
  sum=sum+n
  i=i+1
print(sum)
print(i)
print(sum/i)
```

```python
import numpy
x = numpy.mean(marks)
print(x)
```

#Median
find the middle value of a sorted array

```python
x = numpy.median(marks)
print(x)
```

#Mode
The value that appears the most number of times.

```python
from scipy import stats
x = stats.mode(marks)
print(x)
```

#Standard Deviation?

Standard deviation is a number that describes how spread out the values are.

A low standard deviation means that most of the numbers are close to the mean (average) value.

A high standard deviation means that the values are spread out over a wider range.

```python
import numpy
x = numpy.std(marks)
print(x)
```

#Variance
Indicates how spread out the values are.

square root of the variance is standard deviation.


```python
x=numpy.var(marks)
print(x)
import math
sd=math.sqrt(x)
print(sd)
```

#Percentiles
A number that describes the value that a given percent of the values are lower than.

```python
#An array of the ages of all the people that live in a street.

ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]
print(ages)
```

```python
#What is the age that 75% of the people are younger than?
x = numpy.percentile(ages, 75)
print(x)
```

```python
#What is the age that 25% of the people are younger than?
x=numpy.percentile(ages,25)
print(x)
```
