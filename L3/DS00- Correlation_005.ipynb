#Correlation Charts

### Step 1 - Scatter Chart

Import matplotlib, seaborn and pandas libraries. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

Load the "heart.csv" dataset. Print the first five rows. Every row represents the medical heart health details of one person.

```python
url='https://raw.githubusercontent.com/sharmaroshan/Heart-UCI-Dataset/master/heart.csv'
```

```python
heart = pd.read_csv(url)
heart.head()
```

Replace the columns names in the dataframe with the following labels:


['Age', 'Gender', 'ChestPain', 'RestingBloodPressure', 'Cholestrol', 'FastingBloodSugar', 'RestingECG', 'MaxHeartRateAchivied',
       'ExerciseIndusedAngina', 'Oldpeak', 'Slope', 'MajorVessels', 'Thalassemia', 'Target']

```python
heart.columns = ['Age', 'Gender', 'ChestPain', 'RestingBloodPressure', 'Cholestrol', 'FastingBloodSugar', 'RestingECG', 'MaxHeartRateAchivied',
       'ExerciseIndusedAngina', 'Oldpeak', 'Slope', 'MajorVessels', 'Thalassemia', 'Target']
heart.head()
```

Create a scatter chart with the following two numerical variables : "Age" and "Cholestrol".

```python
fig, ax = plt.subplots(figsize=(7,7))  
ax.scatter(x='Age', y='Cholestrol', data=heart) 
ax.set_xlabel('Age')  
ax.set_ylabel('Cholestrol') 
ax.set_title("Scatter Chart") 
plt.show()
```

Adjust the points colors using the Colestrol value using the "plasma" class-map. Add a colorbar. (tip - use the "c" and "cmap" parameters).

```python
fig, ax = plt.subplots(figsize=(7,7))  
im = ax.scatter(x='Age', y='Cholestrol', c='Cholestrol', cmap='plasma', data=heart) 
fig.colorbar(im, ax=ax)

ax.set_xlabel('Age')  
ax.set_ylabel('Cholestrol') 
ax.set_title("Cholestrol Level vs. Age") 
plt.show()
```

Create a scatter chart with the following two numerical variables : "Age" and "MaxHeartRateAchivied". Color the points using the "MaxHeartRateAchivied".

```python
fig, ax = plt.subplots(figsize=(7,7))  
im = ax.scatter(x='Age', y='MaxHeartRateAchivied', c='MaxHeartRateAchivied', cmap='hot', data=heart) 
fig.colorbar(im, ax=ax)

ax.set_xlabel('Age')  
ax.set_ylabel('MaxHeartRateAchivied') 
ax.set_title("Max Heart Rate Achivied vs Age") 
plt.show()
```

Replace the numerical values in the "Gender" column using the following mapping:
1:'Male', 0:'Female'  

```python
heart['Gender'] = heart['Gender'].map({1:'Male', 0:'Female'})
heart.head()
```

Using Seaborn. create two scatter plots on the same figure vertically stacked. One plot of the "Cholestrol vs Age" colored using "Gender" and the second plot "MaxHeartRateAchivied vs Age" colored using "Gender". Use a large figure size (10,10).

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(10,10))  
sns.scatterplot(x="Age", y="Cholestrol", hue="Gender", data=heart, ax=ax1)
sns.scatterplot(x="Age", y="MaxHeartRateAchivied", hue="Gender", data=heart, ax=ax2)
plt.show()
```

### Step 4 - Hexbin

Load the "weatherAUS" dataset.

```python
url2='https://raw.githubusercontent.com/PacktPublishing/Machine-Learning-for-Absolute-Beginners---Level-3/main/Section%204/weatherAUS.csv'
weatherAUS = pd.read_csv(url2)
weatherAUS.head()
```

Create a scatter chart with the "MinTemp" as x-axis and "MaxTemp" as y-axis. As you can see there are many data points overlapping on each other. A better approach in this case will be to use Hexbin chart. 

```python
fig, ax = plt.subplots(figsize=(10,10))  
sns.scatterplot(x="MinTemp", y="MaxTemp", data=weatherAUS, ax=ax)
plt.show()
```

Create a Hexbin chart using the same x and y data points with a grid size equal to 40. Now we can see that most of the data points are located at the center of the linear line. 

```python
fig, ax = plt.subplots(figsize=(10,10))  
ax.hexbin(x="MinTemp", y="MaxTemp", gridsize=40, cmap= 'inferno', data=weatherAUS )
plt.show()
```

Create two side by side charts on the same figure - one scatter plot and the second hexbin plot. Both of them will use "MinTemp" as x-axis and "Pressure3pm" and y-axis. 

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15,8))  

sns.scatterplot(x="MinTemp", y="Pressure3pm", data=weatherAUS, ax=ax1)
ax2.hexbin(x="MinTemp", y="Pressure3pm", gridsize=40, cmap= 'inferno', data=weatherAUS )

plt.show()
```
